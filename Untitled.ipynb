```python
import math

def calculate_area(radius):
    return math.pi * radius ** 2

calculate_circumference = lambda radius: 2 * math.pi * radius

radius = float(input("Enter the radius of the circle: "))

area = calculate_area(radius)
circumference = calculate_circumference(radius)

print("Area:", round(area, 2))
print("Circumference:", round(circumference, 2))
```

```python
import re

def sum_numbers(numbers):
    return sum(numbers)

text = input("Enter a string: ")

numbers = re.findall(r'\d+', text)
numbers = list(map(int, numbers))

total = sum_numbers(numbers)

double = lambda x: x * 2
doubled_value = double(total)

print("Numbers found:", numbers)
print("Sum:", total)
```

```python
import math

def calculate_squares(numbers):
    square = lambda x: x ** 2
    return list(map(square, numbers))

user_input = input("Enter a list of numbers separated by commas: ")

numbers = list(map(float, user_input.split(",")))

squares = calculate_squares(numbers)

max_square = max(squares)

sum_of_squares = sum(squares)
sqrt_sum = math.sqrt(sum_of_squares)

print("Numbers:", numbers)
print("Squares:", squares)
print("Maximum Square:", max_square)
print("Square Root of Sum of Squares:", sqrt_sum)
```

```python
import re
import math

password = input("Enter a password: ")

has_upper = re.search(r"[A-Z]", password)
has_lower = re.search(r"[a-z]", password)
has_number = re.search(r"[0-9]", password)

if has_upper and has_lower and has_number:
    strength = "Strong"
else:
    strength = "Weak"

def calculate_average(scores):
    avg = lambda nums: sum(nums) / len(nums)
    return avg(scores)

scores_input = input("Enter 5 scores separated by spaces: ")
scores = list(map(float, scores_input.split()))

average = calculate_average(scores)
rounded_average = math.ceil(average)

print("Password strength:", strength)
print("Rounded Average Score:", rounded_average)
```
