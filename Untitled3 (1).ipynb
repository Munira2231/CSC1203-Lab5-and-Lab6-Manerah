```python
numbers = []

for i in range(5):
    num = float(input(f"Enter number {i+1}: "))
    numbers.append(num)

total = sum(numbers)
average = total / len(numbers)
maximum = max(numbers)
minimum = min(numbers)

print("Numbers:", numbers)
print("Sum:", total)
print("Average:", average)
print("Maximum:", maximum)
print("Minimum:", minimum)
```

```python
words = tuple(input(f"Enter word {i+1}: ") for i in range(5))

print("First word:", words[0])
print("Last word:", words[-1])
print("Total number of words:", len(words))

check_word = input("Enter a word to check if it is in the tuple: ")

if check_word in words:
    print(f"'{check_word}' is in the tuple.")
else:
    print(f"'{check_word}' is not in the tuple.")
```

```python
students = {}

for i in range(3):
    name = input(f"Enter student {i+1} name: ")
    score = float(input(f"Enter {name}'s score: "))
    students[name] = score

print("Students and scores:", students)

max_student = max(students, key=students.get)
print("Student with highest score:", max_student, "with score", students[max_student])
```

```python
list1 = list(map(float, input("Enter numbers for first list separated by spaces: ").split()))
list2 = list(map(float, input("Enter numbers for second list separated by spaces: ").split()))

set1 = set(list1)
set2 = set(list2)

print("Union:", set1 | set2)
print("Intersection:", set1 & set2)
print("Difference (set1 - set2):", set1 - set2)
```

```python
names = [input(f"Enter name {i+1}: ") for i in range(5)]

names_tuple = tuple(names)
print("Names tuple:", names_tuple)

name_lengths = {name: len(name) for name in names}
print("Name lengths dictionary:", name_lengths)

unique_lengths = set(name_lengths.values())
print("Unique name lengths set:", unique_lengths)
```
